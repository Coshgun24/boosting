```python
unique_columns=[]
for i in df.columns:
    if(df[i].nunique()<5):
        unique_columns=unique_columns+[i]
```

```python
df.columns=df.columns.str.strip()
```

```python
drop_columns=df[['Liability-Assets Flag','Net Income Flag']]
drop_columns
```

```python
df1=df.drop(columns=['Liability-Assets Flag','Net Income Flag'],axis=1)
df1
```

```python
df1.shape
```

```python
df1.corr()
```

```python
x=df1.iloc[:,1:]
y=df1.loc[:,'Bankrupt']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y ,test_size=0.25,random_state=101,stratify=y)
```

```python
from sklearn import preprocessing
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
import xgboost
from lightgbm import LGBMClassifier
from sklearn.model_selection import train_test_split,GridSearchCV
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Bagging

```python
from sklearn.svm import SVC
svm = SVC(probability=True)
```

```python
bag = BaggingClassifier(base_estimator= svm,
                 n_estimators = 20,
                 max_samples = 0.5,
                 max_features = 8,
                 bootstrap = True,
                 bootstrap_features = True,
                 n_jobs = -1,
                 random_state = 101).fit(x_train,y_train)
```

```python
bag_pred=bag.predict(x_test)
```

```python
accuracy_score(y_test, bag_pred)
```

## Gradient Boost

```python
gbm = GradientBoostingClassifier(learning_rate=0.01,n_estimators=50,verbose=2,random_state = 101).fit(x_train,y_train)
```

```python
gbm_pred=gbm.predict(x_test)
```

```python
accuracy_score(y_test, gbm_pred)
```

## XGBoost

```python
xgbt=XGBClassifier(n_estimators=15,num_leaves=7,max_depth=2,reg_alpha=0.5,reg_lambda=0.7).fit(x_train,y_train)
```

```python
xgbt_pred = xgbt.predict(x_test)
```

```python
accuracy_score(y_test, xgbt_pred)
```

```python
xgbt.score(x_train, y_train)
```

```python
xgbt.score(x_test, y_test)
```
